```python
import pandas as pd
import matplotlib.pyplot as plt
import zipfile
import os

# Шлях до zip-файлу
zip_path = "C:\\Users\\Igor\\Downloads\\homework.zip"#Сюди вставляєте ваш шлях саме до zip файлу
extract_folder = '/new/homework'#Тут можете залишити усе без змі

# Розпакування файл
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_folder)

# Список файлів у розпакованій папці
extracted_files = os.listdir(extract_folder)
extracted_files[0]
```

```python
# Перерахування файлів в папці "data".
data_folder_path = os.path.join(extract_folder, extracted_files[0])
data_folder_path
```

```python

data_files = os.listdir(data_folder_path)
data_files
```

```python
# Функція для завантаження даних із папки та функцій обчислення
def load_and_compute_features(folder_path):
    files = os.listdir(folder_path)
    features_list = []

    for file in files:
        file_path = os.path.join(folder_path, file)
        data = pd.read_csv(file_path)

        # Функції обчислювальної часової області
        features = {
            'mean_X': data['accelerometer_X'].mean(),
            'mean_Y': data['accelerometer_Y'].mean(),
            'mean_Z': data['accelerometer_Z'].mean(),
            'std_X': data['accelerometer_X'].std(),
            'std_Y': data['accelerometer_Y'].std(),
            'std_Z': data['accelerometer_Z'].std(),
            'min_X': data['accelerometer_X'].min(),
            'min_Y': data['accelerometer_Y'].min(),
            'min_Z': data['accelerometer_Z'].min(),
            'max_X': data['accelerometer_X'].max(),
            'max_Y': data['accelerometer_Y'].max(),
            'max_Z': data['accelerometer_Z'].max(),
            'range_X': data['accelerometer_X'].max() - data['accelerometer_X'].min(),
            'range_Y': data['accelerometer_Y'].max() - data['accelerometer_Y'].min(),
            'range_Z': data['accelerometer_Z'].max() - data['accelerometer_Z'].min(),
            'median_X': data['accelerometer_X'].median(),
            'median_Y': data['accelerometer_Y'].median(),
            'median_Z': data['accelerometer_Z'].median(),
            'amplitude_X': data['accelerometer_X'].max() - data['accelerometer_X'].min(),
            'amplitude_Y': data['accelerometer_Y'].max() - data['accelerometer_Y'].min(),
            'amplitude_Z': data['accelerometer_Z'].max() - data['accelerometer_Z'].min(),
        }
        features_list.append(features)
        print(f"len(features_list) = {len(features_list)}")
    return pd.DataFrame(features_list)

# Функції обчислення для кожного типу діяльності
features_idle = load_and_compute_features(os.path.join(data_folder_path, data_files[0]))
features_running = load_and_compute_features(os.path.join(data_folder_path, data_files[1]))
features_stairs = load_and_compute_features(os.path.join(data_folder_path, data_files[2]))
features_walking = load_and_compute_features(os.path.join(data_folder_path, data_files[3]))

# Об’єднання всіх функції в єдиний DataFrame з мітками
features_idle['activity'] = data_files[0]
features_running['activity'] = data_files[1]
features_stairs['activity'] = data_files[2]
features_walking['activity'] = data_files[3]

combined_features = pd.concat([features_idle, features_running, features_stairs, features_walking], ignore_index=True)
#combined_features
features_idle
```

```python
features_stairs
```

```python
# Вибір ознак і цільової змінної
X = combined_features.drop("activity", axis=1)  # Ознаки
y = combined_features["activity"]  # Цільова змінна
X
```

```python
y
```

```python
# Розділення на тренувальний та тестовий набори
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
X_train
```

```python
import time
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Навчання моделі SVM
start_time = time.time()
svm_model = SVC(kernel='linear').fit(X_train, y_train)
svm_training_time = time.time() - start_time
svm_model
```

```python
# Навчання моделі випадкового лісу
start_time = time.time()
rf_model = RandomForestClassifier(n_estimators=100, random_state=42).fit(X_train, y_train)
rf_training_time = time.time() - start_time
rf_model
```

```python
# Оцінка результатів
svm_predictions = svm_model.predict(X_test)
rf_predictions = rf_model.predict(X_test)

# Отримання звіту про класифікацію
svm_report = classification_report(y_test, svm_predictions)
rf_report = classification_report(y_test, rf_predictions)

print("SVM classification report:")
print("Training time:", svm_training_time, "seconds")
print(svm_report)

print("\nRandom Forest classification report:")
print("Training time:", rf_training_time, "seconds")
print(rf_report)
```

Support (Підтримка): Кількість прикладів у кожному класі.
